```python
import requests
import re
from bs4 import BeautifulSoup
import traceback
```

```python
def getHTMLText(url,code='utf-8'):
    try:
        
        r=requests.get(url,timeout=40)
        r.raise_for_status()
        r.encoding=code
        return r.text
    except:
        return ''
    
    
def geturllist(ulist,stockurl):
    html=getHTMLText(stockurl)
    soup=BeautifulSoup(html,'html.parser')#标签数一般用来解析带有标签树，或带有明显的标签的，用来解析
    a=soup.find_all('a')
    
    #find_all函数返回一个list所有的a标签
    for i in a:
        try:
            href=a.attrs['href']
            ulist.append(re.findall(r'[s][hz]\d{6}',href)[0])#[0]的作用？
            #re.find一般针对小范围，纯文本的查找匹配
        except:
            continue#作用为当遇到不满的情况，可以直接忽略，程序不会停止，退出当前循环，自动继续后一循环
def geturlinfo(ulist,stockurl,fpath):
    for stock in ulist:
        url=stockurl+stock+'.html'
        html=getHTMLText(url)
        try:
            infodict={}
            soup=BeautifulSoup(html,'html.parser')
            stockinfo=soup.find('div',attts={'class':'stock-bets'})
            name=stockinfo.find_all(attrs={'class':'bets-name'})[0]
            infodict.update({'股票名称':name.text.spilt()[0]})
            
            keylist=stockinfo.find_all('dt')
            vallist=stockinfo.find_all('dd')
            for i in range(len(keylist)):
                key=keylist[i].text
                val=vallist[i].text
                infodict[key]=val
            with open(fpath,'a',encoding='utf-8') as f:
                f.write(str(infodict)+'\n')
        except:
            traceback.print_exc()
            continue

            


def main():
    stock_list_url='http://quote.eastmoney.com/stocklist.html'
    stock_info_url='http://gupiao.baidu.com/stock/'
    outpath='D:/NewBaiduStockInfo.txt'
    slist=[]
    geturllist(slist,stock_list_url)
    geturlinfo(slist,stock_info_url,outpath)
```

```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""

import requests
import re
from bs4 import BeautifulSoup
import traceback
def getHTMLText(url,code='utf-8'):
    try:
        
        r=requests.get(url,timeout=40)
        r.raise_for_status()
        r.encoding=code
        return r.text
    except:
        return ''
    
def geturllist(stockurl):
    html=getHTMLText(stockurl)
    soup=BeautifulSoup(html,'html.parser')#标签数一般用来解析带有标签树，或带有明显的标签的，用来解析
    lst=soup.body('li')
    Lst=re.findall(r'[s][zh]\d{6}',str(lst))
    return Lst
    
   
    
#def geturlinfo(ulist,stockurl,fpath):
def geturlinfo(ulist,stockurl,fpath):    
    #多个stock，每一个stock对应一个输出
    for stock in ulist:
        url=stockurl+stock+'.html'#构建每一只股票的url
        html=getHTMLText(url)
        try:
            if html=='':#防止在ulist中出现空的行数，continue意味着如果出现空的一行，就中断当前循环继续下一循环
                continue
            infodict={}
            soup=BeautifulSoup(html,'html.parser')
            stockinfo=soup.#返回标签树，tag=div，attrs={class=bets-content}d的全部信息
            name=stockinfo.find_all(attrs={'class':'bets-name'})[0]#name包括的范围？
            infodict.update({'股票名称li' : name.text.split()[0]})
            keylist=stockinfo.find_all('dt')
            valuelist=stockinfo.find_all('dd')
            #每一个stock中包含多个属性keylist,valuelist,为一个list ，把里面的内容一一对应，用一个for循环实现
            for i in range(len(keylist)):
                #·。text用来显示里面的文本内容必须加上text
                key=keylist[i].text
                val=valuelist[i].text
                #dict中直接添加新的列
                infodict[key]=val
            with open(fpath,'a') as f:
                f.write(str(infodict)+'\n')#写入文本，写一项换一行
        except:
            traceback.print_exc()
            continue
        

def main():
    stock_list_url='http://quote.eastmoney.com/stocklist.html'
    stock_info_url='http://gupiao.baidu.com/stock/'
    outpath='D:/BaiduStockInfo.txt'
    slist=geturllist(stock_list_url)
    geturlinfo(slist,stock_info_url,outpath)
main()
```
